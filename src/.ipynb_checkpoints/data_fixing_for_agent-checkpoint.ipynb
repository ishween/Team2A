```python
import pandas as pd
import os
```

```python
data_dir = os.path.join(os.getcwd(), '..', 'data')
```

```python
df = pd.read_csv(os.path.join(data_dir, 'original', 'accounts.csv'))
```

```python
df['account_id'] = [f"ACC-{num:05d}" for num in range(1, len(df) + 1)]
```

```python
df.to_csv(os.path.join(data_dir, 'accounts_with_id.csv'), index=False)
```

```python
opp_df = pd.read_csv(os.path.join(data_dir,'processed', 'original_merged_data.csv'))
```

```python
opp_df
```

```python
opp_df["opportunity_id"] = opp_df["opportunity_id"].map(lambda x: f"OPP-{x}")
```

```python
ids = df[['account_id', 'account']]
ids
```

```python
opp_df = opp_df.merge(ids, on='account', how='left')
```

```python
opp_df.to_csv(os.path.join(data_dir, 'opp_data_with_id.csv'), index=False)
```

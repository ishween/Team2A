```python
import pandas as pd
import os
```

```python
data_dir = os.path.join(os.getcwd(), '..', 'data', 'original')
```

```python
df = pd.read_csv(os.path.join(data_dir, 'accounts.csv'))
```

```python
df['account_id'] = [f"ACC-{num:05d}" for num in range(1, len(df) + 1)]
```

```python
df.to_csv(os.path.join(data_dir, 'accounts_with_id.csv'), index=False)
```
